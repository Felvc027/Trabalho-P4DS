```python
import scipy as sp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
from analyze import analise
import indic
```

```python
ind = indic.indices.split()
lista, var =  [None]*len(ind), 0
pib = analise(2017,'GDP per capita')
for i in ind[:-1]:
    maxi = 0
    for k in range(2015,2018):
        comp = pd.DataFrame(analise(k,i.replace("_", " ").replace(".xls","")))
        comp = comp.join(pib).dropna()
        comp.columns = ["x","y"]
        if len(comp) >= 150:
            reg = sm.ols(formula='y~x', data=comp).fit()
            if reg.rsquared >= maxi:
                maxi = reg.rsquared
                ano = k
        else:
            ano = None
    lista[var] = ano
    var+=1
```

```python
test = pd.DataFrame(pib)
for i in range(len(ind)-1):
    if lista[i] != None:
        a = analise(lista[i],ind[i].replace("_", " ").replace(".xls",""))
        test = test.join(a/(a.mean()))
```

```python
comp = test.dropna()
comp.columns = ["y"] + ['x'+str(i) for i in range(1,len(comp.columns))]
reg = sm.ols(formula='y~x1+x2+x3+x4+x5+x6+x7+x8+x9', data=comp).fit()
reg.summary()
```

```python
positivo, negativo, valp, valn = [],[],[],[]
for i in range(1,len(test.dropna().columns)-1):
    if reg.params[1:][i]>=0:
        valp += [test.columns[i]]
        positivo += [reg.params[1:][i]]
    else:
        valn += [test.columns[i]]
        negativo += [reg.params[1:][i]]
```

```python
valp
```

```python
labels = valp
sizes = positivo
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
labels = valn
sizes = abs(np.array(negativo))
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
